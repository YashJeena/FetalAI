# 6. Find the outliers and replace them outliers

```python
sns.boxplot(data=df, y="price")
```

```python
sns.boxplot(data=df, y="area")
```

```python
Q3 = df["price"].quantile(0.75)
Q1 = df["price"].quantile(0.25)
IQR = Q3 - Q1
IQR
lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR
print(upper_lim,lower_lim)
```

```python
a = np.array(df["price"])
tenth_percentile = np.percentile(a, 5)
ninetieth_percentile = np.percentile(a, 95)
# print(tenth_percentile, ninetieth_percentile)
b = np.where(a<tenth_percentile, tenth_percentile, a)
b = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
df["price"] = b

a = np.array(df["area"])
tenth_percentile = np.percentile(a, 5)
ninetieth_percentile = np.percentile(a, 95)
# print(tenth_percentile, ninetieth_percentile)
b = np.where(a<tenth_percentile, tenth_percentile, a)
b = np.where(b>ninetieth_percentile, ninetieth_percentile, b)
df["area"] = b
```

```python
plt.subplot(1,2,1)
sns.boxplot(data=df, y="price")
plt.title("Price")
plt.subplot(1,2,2)
sns.boxplot(data=df, y="area")
plt.title("Area")
```

# 8. Split the data into dependent and independent variables. 

```python
Y = df["price"]
X = df.drop(["price"], axis=1)
X
```

# 9. Scale the independent variables

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_ = sc.fit_transform(X)
X = pd.DataFrame(X_, columns = X.columns)
# Y = sc.fit_transform(np.array(Y).reshape(-1, 1))
Y
```

# 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

# 11. Build the Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(lr, 5)
rfe = rfe.fit(X_train, Y_train) # Recursive Feature Elimination
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train = X_train[col]
X_train
```

```python
X_test = X_test[col]
```

# 12. Train the Model

```python
lr.fit(X_train, Y_train)
```

# 13. Test the Model


```python
Y_pred = lr.predict(X_test)
```

# 14. Measure the performance using Metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
rs = r2_score(Y_test, Y_pred)
rs
```
