# 5. Handle the Missing values.

### Drop missing values

```python
df = pd.read_csv("titanic.csv")
print("Initially\n", df.isnull().sum() , "\n\n")
df1 = df.dropna()
print("Finally\n", df.isnull().sum() , "\n\n")
print(f"initially - {len(df)}")
print(f"finally - {len(df1)}\n As the length reduces drastically we won't use this method")
```

### Handling missing values(Categorical)

```python
#replace by mode
df["deck"] = df["deck"].fillna(df["deck"].value_counts().index[0])
```

### Handling missing values(Numerical)

```python
#replace by mean
df_clean1 = df["age"].fillna(df["age"].mean())
#replace by median
df_clean2 = df["age"].fillna(df["age"].median())
df["age"] = df_clean1
df
```

```python
df.isnull().sum()
```
